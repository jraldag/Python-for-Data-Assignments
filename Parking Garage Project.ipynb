# Parking Garage Project

Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP). 

Your parking gargage class should have the following methods:
- takeTicket
   - This should decrease the amount of tickets available by 1
   - This should decrease the amount of parkingSpaces available by 1
- payForParking
   - Display an input that waits for an amount from the user and store it in a variable
   - If the payment variable is not empty then ->  display a message to the user that their ticket has been paid and they have 15mins to leave
   - This should update the "currentTicket" dictionary key "paid" to True
-leaveGarage
   - If the ticket has been paid, display a message of "Thank You, have a nice day"
   - If the ticket has not been paid, display an input prompt for payment
      - Once paid, display message "Thank you, have a nice day!"
   - Update parkingSpaces list to increase by 1
   - Update tickets list to increase by 1

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

```python
class Parking_Garage():
    
    def __init__(self, tickets = [0], parkingSpaces = [], currentTicket = {}):
        self.tickets = tickets
        self.parkingSpaces = parkingSpaces
        self.currentTicket = currentTicket
        
    def takeTicket(self):
        if len(self.parkingSpaces) == 5:
            print("The parking garage is currently full.")
        else:
            self.tickets.append(self.tickets[-1] + 1)
            self.parkingSpaces.append(1)
            self.currentTicket.update({str(self.tickets[-1]): 'unpaid'})
            print(f"Please take ticket # {self.tickets[-1]} and proceed with parking your vehicle.")

    def payForParking(self):
        ticket = input("Please enter your ticket #. ")
        if ticket in self.currentTicket:
            i = self.currentTicket[ticket]
            if i == 'unpaid':
                amount = input("Please enter your payment amount. ")
                try:
                    if int(amount) > 0:
                        self.currentTicket.update({ticket: "paid"})
                        self.parkingSpaces.pop()
                        print("Your ticket has been paid. You have 15 minutes to exit the garage.")
                        print("Thank you, have a nice day!")
                    else:
                        print("That is not a valid payment amount.")
                except:
                    print("That is not a valid payment amount.")
            else:
                print("This ticket has been paid.")
                print("Thank you, have a nice day!")
        else:
            print("That is not a valid ticket #.")

        
parking = Parking_Garage()

from IPython.display import clear_output

def park():
    while True:
        response = input("What would you like to do?\n\nYou may:\n- Get a Ticket to Park\n- Pay Ticket for Parking\n- Quit\n").lower()
        if response == 'get a ticket to park':
            clear_output()
            parking.takeTicket()
        elif response == 'pay ticket for parking':
            clear_output()
            parking.payForParking()
        elif response == 'quit':
            clear_output()
            break
        else:
            clear_output()
            print("That is not an available option.")
            
park()  
```
